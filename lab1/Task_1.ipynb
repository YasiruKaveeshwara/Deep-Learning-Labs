```python
#Imports

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import pandas as pd
from scipy.stats import pearsonr, spearmanr
```

```python
# 1. Create a random 4x4 array from exponential distribution

exp_arr_4x4 = np.random.exponential(scale=1.0, size=(4, 4))
print(exp_arr_4x4)
```

```python
# 2. Create a random 100000x1 array from the same exponential distribution and visualize the distribution histogram along with uniform and normal distributions

plt.hist(np.random.exponential(scale=1.0, size=100000), density=True, bins=100, histtype="step", color="Red", label="Exponential")
plt.hist(np.random.rand(100000), density=True, bins=50, histtype="step", color="Green", label="Uniform")
plt.hist(np.random.normal(0, 1, 100000), density=True, bins=25, histtype="step", color="Blue", label="Normal")


plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
# 3. Using matplotlib, plot the Z = X**2 + Y**2 in a 3D plot. Limit the x and y values to [-5, 5]

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

figure = plt.figure(1, figsize = (12, 4))
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', linewidth=0.1)
plt.show()
```

```python
# 4. Calculate the pearson standard coefficient and spearman rank correlation

!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
pokemon_data = df

# Select the stat columns for correlation analysis
stats_columns = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
stats_df = pokemon_data[stats_columns]

# Calculate Pearson and Spearman correlations
pearson_corr = stats_df.corr(method='pearson')
spearman_corr = stats_df.corr(method='spearman')

print("Pearson Correlation Matrix:")
print(pearson_corr)

print("\nSpearman Correlation Matrix:")
print(spearman_corr)

print("\n")

# Heatmaps
plt.figure(figsize=(9,8))
sns.heatmap(pearson_corr, cmap='RdYlBu_r', annot=True)
plt.title("Pearson Heatmap")
plt.show()

plt.figure(figsize=(9,8))
sns.heatmap(spearman_corr, cmap='magma', annot=True)
plt.title("Spearman Heatmap")
plt.show()
```
